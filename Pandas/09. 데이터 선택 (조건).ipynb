#### **9. 데이터 선택 (조건)**
**조건에 해당하는 데이터 선택**

```python
import pandas as pd

df = pd.read_excel('score.xlsx', sheet_name='성적표', index_col='번호')
df.reset_index(drop=True, inplace=True)
df
```

```python
df['키'] >= 185
```

```python
filt = df['키'] >= 185
df[filt]
```

```python
df[~filt] # ~는 not을 의미
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학']
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '사회']]
```

##### **다양한 조건**

###### **& and**

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
```

###### **| or**

```python
df.loc[(df['키'] >= 200) | (df['키'] <= 170)]
```

##### **str 함수**

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('태')
df[filt]
```

```python
df[~filt]
```

```python
langs = ['Python', 'C']
filt = df['SW특기'].isin(langs)
df[filt]
```

```python
langs = ['python', 'c']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt]
```

```python
df['SW특기'].str.contains('Java')
```

```python
df['SW특기'].str.contains('Java', na=False) # NaN을 False로 처리
```

```python
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
