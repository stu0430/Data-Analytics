#### **1. 인구 피라미드**

##### **남자 데이터 정의**

```python
import warnings
import pandas as pd

warnings.filterwarnings(action='ignore')

file_name = '201301_201301_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, index_col='행정기관', usecols='B,E:Y', skiprows=3)
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
df_m.iloc[0]
```

##### **여자 데이터 정의**

```python
df_w = pd.read_excel(file_name, index_col='행정기관', usecols='B,AB:AV', skiprows=3)
df_w.head(3)
```

```python
df_w.columns = df_m.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
df_w.head(3)
```

##### **데이터 시각화**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
plt.figure(figsize=(10, 5))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000, color='#a0c4ff', label='남') # 단위 : 천명
plt.barh(df_w.columns, df_w.iloc[0] // 1000, color='#ffadad', label='여')
plt.title('2013.01 대한민국 인구 피라미드')
plt.legend()
plt.savefig('2013.01 대한민국 인구 피라미드.png', dpi=100)
plt.show()
```
