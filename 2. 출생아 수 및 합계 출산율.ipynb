#### **2. 출생아 수 및 합계 출산율**

```python
import warnings
import pandas as pd

warnings.filterwarnings(action='ignore')

file_name = '142801_20230210195642529_excel.xlsx'
df = pd.read_excel(file_name, skiprows=2, nrows=2, index_col=0)
df
```

```python
df.index
```

```python
df.index.values
```

```python
df.rename(index={'출생아\xa0수':'출생아 수', '합계\xa0출산율' : '합계 출산율'}, inplace=True)
df
```

```python
df.index.values
```

```python
df.loc['출생아 수']
```

```python
df.iloc[0]
```

```python
df = df.T # Transpose
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(13, 5))
ax1.bar(df.index, df['출생아 수'], color='#0AA9F5')
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250, 700)
ax1.set_yticks(range(300, 601, 100))

ax2 = ax1.twinx()
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o', mfc='white', ms=6, lw=2)
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(13, 5))
fig.suptitle('출생아 수 및 합계 출산율', fontsize=15)

ax1.bar(df.index, df['출생아 수'], label='출생아 수', color='#0AA9F5')
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250, 700)
ax1.set_yticks(range(300, 601, 100))
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 10, val, ha='center')
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(df.index, df['합계 출산율'], label='합계 출산율', color='#ffd100', marker='o', mfc='white', ms=6, lw=2)
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.05, val, ha='center')
ax2.legend(loc='upper right')

fig.savefig('출생아 수 및 합계 출산율.png', dpi=100)

plt.show()
```
