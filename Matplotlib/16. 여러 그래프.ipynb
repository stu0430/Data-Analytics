#### **16. 여러 그래프**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
df = pd.read_excel('../Pandas/score.xlsx', sheet_name='성적표', index_col='번호')
df.reset_index(drop=True, inplace=True)
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
plt.suptitle('여러 그래프 넣기')

axs[0, 0].bar(df['이름'], df['국어'], label='국어', width=0.7, color='yellowgreen')
axs[0, 0].set_title('국어 성적')
axs[0, 0].legend()
axs[0, 0].set(xlabel='이름', ylabel='점수')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid('--', linewidth=0.5)

axs[0, 1].plot(df['이름'], df['수학'], label='수학', color='#fd1')
axs[0, 1].plot(df['이름'], df['영어'], label='영어', color='#ffaa11')
axs[0, 1].set_title('수학, 영어 성적')
axs[0, 1].legend()
axs[0, 1].set(xlabel='이름', ylabel='점수')
axs[0, 1].set_facecolor('whitesmoke')
axs[0, 1].grid('--', linewidth=0.5)

axs[1, 0].barh(df['이름'], df['키'], label='키', color='lightcoral')
axs[1, 0].set_title('신장')
axs[1, 0].legend()

axs[1, 0].set(xlabel='키', ylabel='이름')
axs[1, 0].set(xlim=(160, 205))
axs[1, 0].grid('--', linewidth=0.5)

axs[1, 1].plot(df['이름'], df['사회'], label='사회', color='#1cf')
axs[1, 1].set_title('사회 성적')
axs[1, 1].legend()
axs[1, 1].set(xlabel='이름', ylabel='점수')
axs[1, 1].set_facecolor('lightCyan')
axs[1, 1].grid('--', linewidth=0.5)

plt.show()
```
