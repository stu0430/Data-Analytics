#### **7. 여러 데이터**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
x = [1, 2, 3]

y_1 =  [2, 4, 8]
y_2 = [5, 1, 3]
y_3 = [1, 2, 5]

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.show()
```

```python
plt.plot(x, y_1, label='data 1')
plt.plot(x, y_2, 'o--', label='data 2')
plt.plot(x, y_3 , 's-.', label='data 3')

plt.legend()
plt.show()
```

```python
plt.plot(x, y_1, label='data 1')
plt.plot(x, y_2, 'o--', label='data 2')
plt.plot(x, y_3 , 's-.', label='data 3')

plt.legend(ncol=3)
plt.show()
```
