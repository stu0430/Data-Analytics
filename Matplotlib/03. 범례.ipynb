#### **3. 범례 (legend)**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 8, 16, 32]
```

```python
plt.plot(x, y, label='데이터')
plt.legend()
plt.show()
```

```python
plt.plot(x, y, label='데이터')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.plot(x, y, label='데이터')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(x, y, label='데이터')
plt.legend(loc=(0.5, 0.5)) # x, y 좌표 (0 ~ 1)
plt.show()
```
