#### **2. 축**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 8, 16, 32]
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프', fontdict={'family' : 'HYGungSo-Bold', 'size': 20})
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color='#00aacc')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red', loc='right')
plt.ylabel('Y축', color='#00aacc', loc='top')
plt.show()
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([10, 20, 30, 40, 50])
plt.show()
```
