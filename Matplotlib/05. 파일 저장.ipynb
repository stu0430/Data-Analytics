#### **5. 파일 저장**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 8, 16, 32]
```

```python
plt.plot(x, y)

plt.savefig('graph.png', dpi=100)
```

```python
plt.figure(dpi=200)
plt.plot(x, y)
plt.savefig('graph_200.png', dpi=100)
```
