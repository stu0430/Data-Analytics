#### **11. 누적 막대 그래프**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
df = pd.read_excel('../Pandas/score.xlsx', sheet_name='성적표', index_col='번호')
df.reset_index(drop=True, inplace=True)
df
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'], df['영어'])
plt.show()
```

```python
plt.bar(df['이름'], df['국어'], width=0.5)
plt.bar(df['이름'], df['영어'], bottom=df['국어'], width=0.5)
plt.show()
```

```python
plt.bar(df['이름'], df['국어'], width=0.5, label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], width=0.5, label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'], width=0.5, label='수학')

plt.xticks(rotation=45)

plt.legend()
plt.show()
```
