#### **4. 스타일**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 8, 16, 32]
```

```python
plt.plot(x, y, linewidth=3)
plt.show()
```

##### **마커**

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o', linestyle='None')
plt.show()
```

```python
plt.plot(x, y, marker='v')
plt.show()
```

```python
plt.plot(x, y, marker='v', markersize=10)
plt.show()
```

```python
plt.plot(x, y, marker='X', markersize=10)
plt.show()
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red')
plt.show()
```

```python
plt.plot(x, y, marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow')
plt.show()
```

##### **선 스타일**

```python
plt.plot(x, y, linestyle=':')
plt.show()
```

```python
plt.plot(x, y, linestyle='--')
plt.show()
```

```python
plt.plot(x, y, linestyle='-.')
plt.show()
```

```python
plt.plot(x, y, linestyle='-')
plt.show()
```

##### **색깔**

```python
plt.plot(x, y, color='pink')
plt.show()
```

```python
plt.plot(x, y, color='#ff0000')
plt.show()
```

```python
plt.plot(x, y, color='b')
plt.show()
```

```python
plt.plot(x, y, color='g')
plt.show()
```

##### **포맷**

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
plt.show()
```

```python
plt.plot(x, y, 'bv:')
plt.show()
```

```python
plt.plot(x, y, 'go')
plt.show()
```

##### **축약어**

```python
plt.plot(x, y, marker='o', mfc='red', mec= 'blue', ms=10, ls=':')
plt.show()
```

##### **투명도**

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.5)
plt.show()
```

##### **그래프 크기**

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(8, 4), dpi=50)
plt.plot(x, y)
plt.show()
```

##### **배경색**

```python
plt.figure(facecolor='lightgray')
plt.plot(x, y)
plt.show()
```

```python
plt.figure(facecolor='#bbee')
plt.plot(x, y)
plt.show()
```
