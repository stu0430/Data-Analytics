#### **13. 원 그래프 (기본)**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic', size=10)
plt.rc('axes', unicode_minus=False)
plt.rc('figure', figsize=(4.5, 3))
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'Rust', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.03] * 6

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)

plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도')

plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```
