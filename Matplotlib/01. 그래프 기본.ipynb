#### **Matplotlib**
**다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### **1. 그래프 기본**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 8, 16, 32]

plt.plot(x, y)

plt.show()
```

##### **Title 설정**

```python
plt.plot(x, y)
plt.title('Line Graph')
plt.show()
```
